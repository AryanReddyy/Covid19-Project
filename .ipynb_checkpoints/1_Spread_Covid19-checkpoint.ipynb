```python
import matplotlib.pyplot as plt

# Aggregate the data to state level by summing up all the counties within each state
confirmed_state_df = confirmed_df.groupby('Province_State').sum()
deaths_state_df = deaths_df.groupby('Province_State').sum()

# Drop the non-numeric columns that are now irrelevant after aggregation
confirmed_state_df = confirmed_state_df.drop(columns=['UID', 'code3', 'FIPS', 'Lat', 'Long_'])
deaths_state_df = deaths_state_df.drop(columns=['UID', 'code3', 'FIPS', 'Lat', 'Long_', 'Population'])

# Transpose the dataframes to have dates as rows and states as columns
confirmed_state_transposed = confirmed_state_df.transpose()
deaths_state_transposed = deaths_state_df.transpose()

# Plot the total confirmed cases and deaths over time for the US
plt.figure(figsize=(15, 10))

plt.subplot(2, 1, 1)
plt.title('Total COVID-19 Confirmed Cases in the US')
plt.xlabel('Date')
plt.ylabel('Number of Confirmed Cases')
plt.plot(confirmed_state_transposed.sum(axis=1), label='Total Confirmed Cases')
plt.legend()

plt.subplot(2, 1, 2)
plt.title('Total COVID-19 Deaths in the US')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.plot(deaths_state_transposed.sum(axis=1), label='Total Deaths', color='red')
plt.legend()

plt.tight_layout()
plt.show()
```

```python
# Convert 'date' to datetime format and sort the data
vaccinations_df['date'] = pd.to_datetime(vaccinations_df['date'])
vaccinations_df = vaccinations_df.sort_values(by='date')

# Aggregate the data to get total vaccinations, people vaccinated, and people fully vaccinated for the US
vaccinations_aggregated_df = vaccinations_df.groupby('date').sum()

# Plot the vaccination data
plt.figure(figsize=(15, 7))

plt.title('COVID-19 Vaccination Progress in the US')
plt.xlabel('Date')
plt.ylabel('Number of People')
plt.plot(vaccinations_aggregated_df['people_vaccinated'], label='People Vaccinated', color='blue')
plt.plot(vaccinations_aggregated_df['people_fully_vaccinated'], label='People Fully Vaccinated', color='green')

plt.legend()
plt.show()
```

```python
import plotly.graph_objects as go

# Prepare the data for plotting
# Aggregate confirmed cases and deaths data to country level by summing over all dates
confirmed_total = confirmed_state_df.sum(axis=0)
deaths_total = deaths_state_df.sum(axis=0)

# Create date index for confirmed and deaths (same as columns in the transposed DFs)
dates = confirmed_state_df.columns

# For vaccinations, use the aggregated data we prepared earlier
vaccinated_total = vaccinations_aggregated_df['people_vaccinated']

# Create the Plotly figure
fig = go.Figure()

# Add confirmed cases trace
fig.add_trace(go.Scatter(x=dates, y=confirmed_total, name='Confirmed Cases',
                         line=dict(color='blue'), mode='lines'))

# Add deaths trace
fig.add_trace(go.Scatter(x=dates, y=deaths_total, name='Deaths',
                         line=dict(color='red'), mode='lines'))

# Add vaccinations trace, aligning the dates
vaccinated_dates = vaccinated_total.index
vaccinated_values = vaccinated_total.values
fig.add_trace(go.Scatter(x=vaccinated_dates, y=vaccinated_values, name='Vaccinated',
                         line=dict(color='green'), mode='lines'))

# Set plot layout
fig.update_layout(title='COVID-19 Confirmed Cases, Deaths, and Vaccinations Over Time in the US',
                  xaxis_title='Date',
                  yaxis_title='Count',
                  legend_title='Legend')

# Show the figure
fig.show()
```
